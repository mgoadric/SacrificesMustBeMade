```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import math
```

```python
speed1 = 0.08
speed2 = 0.082
r1 = 0.003
r2 = 0.004

sim_length = 60

def euler(func, initial, deltas):
    vals = [initial]
    for i in range(1, len(deltas)):
        dt = deltas[i] - deltas[i - 1]
        initial += func(initial, dt)
        vals.append(initial)
    return vals
    
def growth(u, dt):
    return u * r1 * dt

def growth2(u, dt):
    return u * r2 * dt

xs = np.linspace(0, sim_length, sim_length * 2 + 1)
ys1 = euler(growth, speed1, xs)
ys2 = euler(growth2, speed2, xs)

plt.plot(xs, ys1)
plt.plot(xs, ys2)
plt.plot(xs, [.1] * 121)
plt.legend(["Friend", "Enemies", "You"])
plt.ylabel("Speed")
plt.xlabel("Seconds")
plt.title("Speed of characters")
```

```python
def euler2(initial, speeds):
    vals = [initial]
    for i in range(1, len(speeds)):
        initial += 30 * speeds[i]
        vals.append(initial)
    return vals
```

```python
yd1 = euler2(-4.7, ys1)
yd2 = euler2(-15.5, ys2)
ydc = euler2(-5, [0] * 4 + [.1] * 117)

plt.plot(xs, yd1)
plt.plot(xs, yd2)
plt.legend(["Friend", "Enemies"])
plt.ylabel("Distance")
plt.xlabel("Seconds")
plt.title("Distance Travelled")
```
